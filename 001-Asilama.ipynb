```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataFrame = pd.read_csv("vaccination-data.csv")
```

```python
dataFrame
```

```python
avrupa = dataFrame[dataFrame["WHO_REGION"] == "EURO"]
```

```python
avrupa
```

```python
filtrelenmis_avrupa = avrupa[["ISO3","TOTAL_VACCINATIONS"]]
```

```python
filtrelenmis_avrupa = filtrelenmis_avrupa.rename(columns={"ISO3":"ISO3_CODE"})
```

```python
filtrelenmis_avrupa
```

```python
fig, ax = plt.subplots(figsize =(10,6))# Figure ve axes oluÅŸturun
# DataFrame'i tablo olarak gÃ¶sterin
table = plt.table(cellText=filtrelenmis_avrupa.values,# DataFrame'in deÄŸerlerini kullanÄ±n
                 colLabels=filtrelenmis_avrupa.columns,# SÃ¼tun baÅŸlÄ±klarÄ±nÄ± kullanÄ±n
                 cellLoc='center',# HÃ¼cre iÃ§eriÄŸini ortala
                 loc='center')# Tabloyu ortala
# Tabloyu dÃ¼zenleme
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2,1.2)# Tablo boyutunu ayarlayÄ±n

ax.axis('off')# Eksenleri ve Ã§erÃ§eveyi gizle

```

```python
fig, ax = plt.subplots(figsize=(10,6)) # Figure ve axes oluÅŸturun
x = filtrelenmis_avrupa["ISO3_CODE"] # x eksenine Ã¼lke isimlerini alÄ±n
y = filtrelenmis_avrupa["TOTAL_VACCINATIONS"] # y eksenine toplam aÅŸÄ±lama verilerini alÄ±n

ax.bar(x,y) # Ã‡ubuk grafik oluÅŸturun
# Eksen ve baÅŸlÄ±k etiketlerini ekleyin
ax.set_xlabel("Ãœlkeler")
ax.set_ylabel("Toplam AÅŸÄ±lama(milyon)")
ax.set_title("Euro BÃ¶lgesi Ãœlkelerde Toplam AÅŸÄ±lama")
# x ekseni etiketlerini dÃ¶ndÃ¼rÃ¼n ve sÄ±kÄ±ÅŸtÄ±rÄ±n
plt.xticks(rotation=90)
plt.tight_layout()
```
